```python
def are_equal(val1, val2, val3, val4):
  return val1 == val2 == val3 == val4
```

```python
idx, correct_label_same_pred, incorrect_label_same_pred = 0, [], []
for (label1, label2, label3, label4) in zip(log_df["Predicted Label"], nb_df["Predicted Label"], gru_df["Predicted Label"], transformer_df["Predicted Label"]):
  if are_equal(label1, label2, label3, label4):
    if log_df["Predicted Label"].iloc[idx] == log_df["True Label"].iloc[idx]:
      correct_label_same_pred.append(idx)
    else: incorrect_label_same_pred.append(idx)
  idx += 1
```

```python
def map_label_to_class(indices, classes):
  return [classes[idx] for idx in indices]
```

```python
def create_csv(correct_label_same_pred, incorrect_label_same_pred, file_path="same_predictions.csv"):
  with open(file_path, 'w') as file:
    file.write('Examples correctly predicted accross all models\n')
    file.write('Index,Label\n')
    for example in correct_label_same_pred:
      file.write(f'{example},{log_df["Predicted Label"].iloc[example]}\n')
    file.write('Examples incorrectly predicted accross all models, but the models predicted same label\n')
    file.write('Index,Label\n')
    for example in incorrect_label_same_pred:
      file.write(f'{example},{log_df["Predicted Label"].iloc[example]}\n')
  print("Content has been written to the file.")
```

```python
create_csv(correct_label_same_pred, incorrect_label_same_pred)
```
